```python
# Webscrapping

from bs4 import BeautifulSoup
import requests


url = "https://en.wikipedia.org/wiki/2020_coronavirus_pandemic_in_Indonesia"
html = requests.get(url)
soup = BeautifulSoup(html.content, "html.parser")

table = soup.find(class_="wikitable float sortable")

th = []
for i in table.find_all('th'):
    th.append(i.text[:-1])
    
label = []
for i in range(0,5):
    label.append(th[i])

prov = []
for i in range(11,45):
    prov.append(th[i])
prov[prov.index('Jakarta[b]')]= 'Jakarta'

data = []
for i in table.find_all('td'):
    data.append(i.text)

con = []
indexcon = list(range(0,len(data),5))
recover = []
indexre = list(range(1,len(data),5))
death = []
indexd = list(range(2,len(data),5))
active = []
indexa = list(range(3,len(data),5))


for i in range(len(data)):
    if i in indexcon:
        con.append(data[i][:-1])
    elif i in indexre:
        recover.append(data[i][:-1])
    elif i in indexd:
        death.append(data[i][:-1])
    elif i in indexa:
        active.append(data[i][:-1])

datacovid = []
for i in range(len(prov)):
    datacovid.append({
        label[0] : prov[i],
        label[1] : con[i],
        label[2] : recover[i],
        label[3] : death[i],
        label[4] : active[i],
    })
```

```python
# Get request data location

import requests

url = "https://raw.githubusercontent.com/LintangWisesa/Indonesia-Covid19-Maps/master/data/gps_indonesia.json"
data1 = requests.get(url)

dataloc = data1.json()
dataloc
```

```python
import folium as fo

covidmap = fo.Map(
    location=[-0.023,116.192],
    titles='OpenStreetMap',
    zoom_start=5
)


for i in range(len(dataloc)):
    fo.Marker(
        [dataloc[i]['latitude'], dataloc[i]['longitude']],
        popup=datacovid[i]['Province']+'\n'+'Confirmed: '+datacovid[i]['Confirmed']+'\n'+'Recovered: '+datacovid[i]['Recovered']+'\n'+'Deaths: '+datacovid[i]['Deaths'],
        tooltip=dataloc[i]['Provinsi']
    ).add_to(covidmap)
    
covidmap.save('covidmap.html')

covidmap
```
